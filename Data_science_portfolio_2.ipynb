**Data Processing**

  Using data processing methods, I will perform functions on my selected columns that will be used in my data visualisations.

```python
df = df.dropna(subset = ['Sex', 'COVID-19 Deaths','Year','Age Group','Total Deaths','Month', 'State']) #Disregarding the entire row if values are missing
df
```

For the selected variables I want to use in my visualisations. I will use my categorical data (months) and convert them to numerical data for tidier semantics.

```python
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12}
```

I have also defined the function of categorizing the age group into 3 sub catorgories (children, adults,  and seniors) so when it comes to displaying multiple values in hue, the data becomes more refined.

```python
def categorize_age_group(age_group):
  if age_group in ['Under 1 year', '1-4 years', '5-14 years', '0-17 years']:
    return 'children'
  elif age_group in ['18-29 years', '30-39 years', '40-49 years', '50-64 years']:
        return 'Adults'
  elif age_group in ['65-74 years', '75-84 years', '85 years and over']:
        return 'Seniors'
```

```python
filtered_df = df[df['Sex'] != 'All Sexes'] #filtering out data where 'sex' is 'all sexes' to make data more refined
```

```python
sns.scatterplot(data=filtered_df, x='COVID-19 Deaths', y='Total Deaths', hue='Age Group', alpha=0.7, palette='colorblind')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Age Group') #moving the legend outside of the dataframe for clear viewing
```

Total fatalities and COVID-19 deaths have a definite positive association, which means that when COVID-19 deaths rise, so do overall deaths. We can also see that higher total death rates appear to be linked to older age groups (e.g., 65–74 years, 75–84 years, and 85 years and above, suggesting that they are more vulnerable to contraction and higher mortality risks.  

(Wilke, 2019) discusses how when we want to visualise more than 2 quantitative variables like the scatterplot above, where the 2 variables are in the same quantity but different conditions. we can add further variables like style and size to differentiate the data as shown in the scatterplot below.  

Jaques Bertin (1967) concept of visual variables, his "retinal" values include the shape, position, size, lightness(value), orientation, color hue and texture. I will apply some of these values so we can add further variables like style and size to differentiate the data as shown in the scatterplot below  

```python
sns.scatterplot(data=filtered_df, x='COVID-19 Deaths', y='Total Deaths', hue='Age Group',size='Sex',
style='Month', alpha=0.7, palette='colorblind')   #Using aesthetics like style and size to differentiate my data
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Age Group')
```

The colour indicates the people's age range. It is simple to visually differentiate data points by age because each age group has a unique colour. Furthermore, the style marker shape represents the months (1-12) and the size markers differentiates sex (male and female).

Using our style semantics, we can visualise the mortality trends if they get categorised into certain months, potentially understanding correlations from seasonal spikes, particularly in winter when the prevalence of respiratory illness is higher.


Choosing a smaller dataset to perform different visualisation https://github.com/mwaskom/seaborn-data/blob/master/planets.csv [Accessed 6 Dec. 2024]. I will perform a scatterplot visualualisation with a logarithmic scale allowing me to cover wider data whilst maintaining clarity in the smaller values

I will import Plotly as px. Plotly is an open-source, python library that providing 40 different chart chart types for a variety of 3- dimensional use cases(plotly.com). This provides us with dynamic and interactive visualisation techniques.

```python
import plotly.express as px
```

```python

df = sns.load_dataset('planets')
df = df.dropna() #removing any NaN values
```

```python
g = sns.scatterplot(data=df, x='orbital_period', y='year', hue='method')
g.set_title('Scatterplot of Orbital Period vs. Year (Log-Log Scale)', fontsize=16)
g.set_xscale('log')  #setting logarithmic scale, refining the data that spreads over multiple orders of magnitude
g.set_yscale('log')
g.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Method') #moving legend outside of plot for better readibility
```

The orbital period ranges from around one day to over 10,000 days, which is several orders of magnitude. Using the logarithmic scale has helped me visualise the large range of data into a smaller plot.

The short orbital ranges are all densley packed, suggesting that objects  (such as celestial bodies, stars, or exoplanets) have short orbital periods.

There are not many larger orbital periods (over 100), which highlights that there are less planets with long cycles which could be (far-off stars, outer planets, or elliptical-orbiting objects).



**Animation Visualisations**

(Taylor, 2022) discussed the importance of animated data visualisations, noting that  "engagement is key because a data visualisation that fails to engage users also fails to convey any information". Below is a animated parrallel coordinate plot of the distance and orbital period over different years.

```python
px.sunburst(df, path=['distance', 'method'],
values='year', color='orbital_period')
```

Using Plotly express, I plotted a hierarchal sunburst plot that shows the relationship between the dimensions(distance and method). With the inner most circle representing distance and the second level corresponding to the method.

The colour mapping used represents the oribital period of 17337.5 days, with most orbital periods below 2k , with the yellow being marked in the upper scale of orbital period, suggesting the orbital time was a lot longer than the rest of the planets.

Parallel Coordinates diagrams allow us to see Multiple dimensions (number, orbital_period, mass, distance, year, method_id, mass_id) plotted together,  showing their correlations and interactions.
```python
px.treemap(df, path=['orbital_period', 'method'],
values='year', color='distance')
```

The Treemap visualisation helps with the handling of bigger hierarchies. The hierarchies are represented through rectangles, making it easier to handle multiple levels of data as it utilizes the availible space. Most of the  rectangles have a distance below 200, this indiocates a concentration of lower values

Treemaps are a better way of presenting flat data, as sunbursts are encoded using angular width and parrallel coordinates diagram can become cluttered with many lines, which makes visual comparisons more difficult and less natural for smaller proportions in the planets dataset.
