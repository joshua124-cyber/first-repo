**Using Seaborn to plot data**

Seaborn is a robust Matplotlib-based Python data visualisation library. It is especially useful for visualising complicated datasets and offers a high-level interface for creating eye-catching statistical graphics.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
sns.set_style("whitegrid")  #Adds a clean grid style
```

```python
plt.grid(axis='y',linestyle='--',color='gray', alpha=0.7)
```

```python
sns.barplot(data=df, x='Year', y='COVID-19 Deaths', color='blue')
plt.title('COVID-19 Deaths by Year', fontsize=14, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('COVID-19 Deaths', fontsize=12)
plt.grid(axis='y', linestyle='--', color='gray', alpha=0.7)
plt.grid(axis='x', linestyle='--', color='lightgray', alpha=0.5)
max_deaths = df['COVID-19 Deaths'].max()
plt.xticks(rotation=45, fontsize=10)
```

The "COVID-19 Deaths by Year" bar graph displays the total number of COVID-19 and pneumonia-related deaths from 2020 to 2023.

The Y-axis, which ranges from 0 to 400, shows the number of COVID-19 and pneumonia-related deaths. The years 2020, 2021, 2022, and 2023 are represented by the X-axis.

in 2020 there was approximatly  deaths. in 2021, the highest number of deaths was recorded with around 220, however there is a notable error bar which suggest variability or uncertainty in the data. 2022 deaths started to show a decline and more noticibly in 2023 the death rate for pneumonia and COVID-19 with a significant drop.

```python
sns.barplot(data=df, hue='Sex', y='COVID-19 Deaths', estimator=sum, palette='viridis')
plt.title('Total COVID-19 Deaths by Sex', fontsize=14, weight='bold')
plt.ylabel('COVID-19 Deaths', fontsize=12)
plt.ylim(0, 20000000)
for container in plt.gca().containers:plt.bar_label(container, fmt='%.0f', label_type='edge', fontsize=10)
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=11) #using legend to move sex outside of the barplot
plt.tight_layout()
plt.show()
```

Total COVID-19 deaths by sex are displayed in this bar chart in the following three categories: "All Sexes," "Male," and "Female." The total number of deaths is shown on the y-axis; the units are probably in tens of millions.

With 8.56 million male deaths and 6.81 million female deaths, the overall death toll was higher for males than females. This discrepancy could be the result of societal influences influencing COVID-19 outcomes, biological variations, or pre-existing health problems.

Male deaths constitute about 55% of total deaths (8.56 million out of 15.42 million), while female deaths make up about 45%.

**Using Hue with bar plots**

By showing distinct bars within each category, the hue parameter in bar plots allows data to be distinguished depending on a categorical variable. It makes it simpler to compare several variables or conditions within each category by enabling the division of colour bars in a dataset based on a particular property.

Seaborn uses hue to differentiate across several data categories. In this case i want to use 'sex' as hue to add a different varible to the data, producing me a more refined barplot.

```python
sns.barplot(data=df, x='Year', y='COVID-19 Deaths', hue='Sex', palette='viridis')
plt.title('COVID-19 Deaths by Year', fontsize=14, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('COVID-19 Deaths', fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=11)

```

Following the analysis made from the table above, Hue has now seperated the death rate by year into male and female and all sexes.

Because of irregular data reporting early in the pandemic, the error bars are bigger in the earlier years (particularly 2020 and 2021), suggesting greater unpredictability or uncertainty in death estimates.

Some clear observations that hue has presented us with is that male death rates remained higher than female death rates over the course of the pandemic.

After reaching a peak in 2021, the overall number of deaths for all sexes steadily decreased in 2022 and 2023. This trend is consistent with improved therapies, and eventually less severe variations.

```python
sns.set_theme(font_scale=1.5, style='white') # Set Seaborn theme with larger font scale and dark style
```

```python
plt.figure(figsize=(8, 6)) # Create a plot with a legend (e.g., using hue to create separate legend items)
ax = sns.scatterplot(data=df, x='Year', y='COVID-19 Deaths', hue='Sex', palette='viridis')
sns.move_legend(ax, loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45, fontsize=12)
```

Scatterplots allow each individual data point to be represented separately, which is useful for identifying specific fluctuations within a time series. This level of granularity is harder to achieve with a bar plot, which typically averages data across each category or time period.

each scatter point is a single data entry that displays the number of deaths for a certain sex group for a given period of time. We can observe the changes in death rates over time and between the sex groups thanks to the distribution of points over the years.



```python
sns.pointplot(data=df, x='Year', y='COVID-19 Deaths', hue='Sex',
              markers='o', linestyle='none', palette='viridis')
plt.title('COVID-19 Deaths by Year and Sex', fontsize=14, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('COVID-19 Deaths', fontsize=12)
plt.tight_layout()
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=11)
```

This pointplot has a single point per year catergory, which most likely represents the average or cumulative number of fatalities for that year.

The point plot makes it easy to see the pattern over time because each year's summary is distinct and easy to understand. By displaying the variation within each year, which the scatterplot does naturally but less explicitly, the vertical lines add another layer.


```python
sns.histplot(data=df, x='Year', hue='Sex', multiple='stack', palette='viridis' , kde=False)
plt.title('Distribution of Entries by Year and Sex', fontsize=14, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('COVID-19 Deaths', fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

```
